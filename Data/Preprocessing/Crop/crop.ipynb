```python
# í°ìƒ‰ ë°°ê²½ 

import cv2
import numpy as np
import os
from PIL import Image

# âœ… ì…ë ¥ ë° ì¶œë ¥ í´ë” ì„¤ì •
input_folder = r"C:\Users\EL0021\Desktop\MS AI School\project-1\data\vegetable_13class\fresh\Cabbage"
output_folder = r"C:\Users\EL0021\Desktop\MS AI School\project-1\data\vegetable_13class\vegetable_v2\Fresh\Cabbage"

# âœ… ì¶œë ¥ í´ë” ì—†ìœ¼ë©´ ìƒì„±
if not os.path.exists(output_folder):
    os.makedirs(output_folder)

# âœ… c_ë¡œ ì‹œì‘í•˜ëŠ” ì´ë¯¸ì§€ë§Œ í•„í„°ë§
image_files = [f for f in os.listdir(input_folder) if f.endswith((".png", ".jpg", ".jpeg"))]

for file_name in image_files:
    img_path = os.path.join(input_folder, file_name)

    # âœ… ì´ë¯¸ì§€ ë¡œë“œ
    image = cv2.imdecode(np.fromfile(img_path, dtype=np.uint8), cv2.IMREAD_COLOR)

    if image is None:
        print(f"âŒ Error loading image: {file_name}")
        continue

    # âœ… GrabCutì„ ìœ„í•œ ì´ˆê¸° ë§ˆìŠ¤í¬ ìƒì„±
    mask = np.zeros(image.shape[:2], np.uint8)

    # âœ… ë°°ê²½ê³¼ ì „ê²½ì„ ëª¨ë¸ë§í•  ë°°ì—´ ìƒì„±
    bgd_model = np.zeros((1, 65), np.float64)
    fgd_model = np.zeros((1, 65), np.float64)

    # âœ… ROI ì„¤ì • (ì „ì²´ ì´ë¯¸ì§€ì—ì„œ ì‹¤í–‰)
    rect = (10, 10, image.shape[1] - 20, image.shape[0] - 20)  # ê²½ê³„ë¥¼ ì¡°ê¸ˆ ë‚¨ê²¨ì„œ ì¡ìŒ ë°©ì§€

    # âœ… GrabCut ì‹¤í–‰
    cv2.grabCut(image, mask, rect, bgd_model, fgd_model, 5, cv2.GC_INIT_WITH_RECT)

    # âœ… ë°°ê²½ ì œê±° (0ê³¼ 2ëŠ” ë°°ê²½, 1ê³¼ 3ì€ ì „ê²½)
    mask2 = np.where((mask == 2) | (mask == 0), 0, 1).astype("uint8")
    image_fg = image * mask2[:, :, np.newaxis]

    # âœ… ë°°ê²½ì„ í°ìƒ‰ìœ¼ë¡œ ë³€ê²½
    white_bg = np.ones_like(image) * 255  # í°ìƒ‰ ë°°ê²½ ìƒì„±
    final_image = white_bg * (1 - mask2[:, :, np.newaxis]) + image_fg

    # âœ… ë¦¬ì‚¬ì´ì¦ˆ (600x600, ê³ í•´ìƒë„ ë³´ê°„)
    resized_image = cv2.resize(final_image.astype(np.uint8), (600, 600), interpolation=cv2.INTER_CUBIC)

    # âœ… ì €ì¥ ê²½ë¡œ ì„¤ì •
    output_path = os.path.join(output_folder, file_name)

    # âœ… OpenCVë¡œ ì €ì¥
    cv2.imencode('.png', resized_image)[1].tofile(output_path)

    # âœ… PILì„ ì´ìš©í•˜ì—¬ DPI ì„¤ì • í›„ ì €ì¥ (ì„ íƒ ì‚¬í•­)
    pil_img = Image.fromarray(cv2.cvtColor(resized_image, cv2.COLOR_BGR2RGB))
    pil_img.save(output_path, dpi=(300, 300))  # 300 DPI ì„¤ì •

    print(f"âœ… Processed and saved: {output_path}")

print("ğŸ‰ ëª¨ë“  ì´ë¯¸ì§€ ì²˜ë¦¬ ì™„ë£Œ!")
```
